# Creating DB for exploratory data analysis

## Setting up common parameters

```python
import sqlalchemy as sqla
import pandas as pd
```

```python
db_file_path = "/Users/kschan/Documents/Education/Advanced Professional Certificate in Data Science and AI/Project/personal/db/world-athletics-performances.db"
engine = sqla.create_engine(f'duckdb:///{db_file_path}')
```

```python
csv_folder_path = "/Users/kschan/Documents/Education/Advanced Professional Certificate in Data Science and AI/Project/personal/data/"
```

## Importing all_disciplines_combined

```python
csv_file_path_all_disciplines_combined = f"{csv_folder_path}all_disciplines_combined.csv"
df_all_disciplines_combined = pd.read_csv(csv_file_path_all_disciplines_combined)
df_all_disciplines_combined.describe()
```

```python
df_all_disciplines_combined.head()
```

```python
df_all_disciplines_combined.to_sql('all_disciplines_combined', con=engine, if_exists='replace', index=True)
```

## Importing individual_events

```python
csv_file_path_individual_events = f"{csv_folder_path}individual_events.csv"
df_individual_events = pd.read_csv(csv_file_path_individual_events)
df_individual_events.describe()
```

```python
df_individual_events.head()
```

```python
df_individual_events.to_sql('individual_events', con=engine, if_exists='replace', index=True)
```

## Importing relay_events

```python
csv_file_path_relay_events = f"{csv_folder_path}relay_events.csv"
df_relay_events = pd.read_csv(csv_file_path_relay_events)
df_relay_events.describe()
```

```python
df_relay_events.head()
```

```python
df_relay_events.to_sql('relay_events', con=engine, if_exists='replace', index=True)
```
