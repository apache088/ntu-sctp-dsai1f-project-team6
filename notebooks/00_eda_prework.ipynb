```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from Covid Data.csv to df
df = pd.read_csv('../data/all_disciplines_combined.csv')

df.head()

# Count of all duplicate rows (keeping all occurrences)
duplicate_rows = df[df.duplicated(keep=False)]
duplicate_count = len(duplicate_rows)

# Count of unique rows (non-duplicates)
unique_rows = df.drop_duplicates()
unique_count = len(unique_rows)

print(f"Total rows with duplicates: {duplicate_count}")
print(f"Total unique rows: {unique_count}")
print(f"Total original rows: {len(df)}")
print(f"Total original rows: {len(df)}")
```
